```python
# Parameters
monthly_interest_rate=0.015 #Monthly interest rate (1.5%)
monthly_payment=50 #Fixed monthly payment ($)

#Calculate equilibrium value (if only interest is paid)
# Parameters
monthly_interest_rate=0.015 #Monthly interest rate (1.5%)
monthly_payment=50 #Fixed monthly payment ($)

#Calculate equilibrium value (if only interest is paid)
equilibrium_value=monthly_payment/monthly_interest_rate

#Display equilibrium value
print(f"Equilibrium Value: ${equilibrium_value:.2f}")

#Simulate balance reduction over time
initial_debt = 500 #initial credit card balance

balance=initial_debt
month = 11

print("\nMonths | Balance ($)")
print("-----------------")
while balance > 0:
      #Calculate interest and update balance
       interest = balance * monthly_interest_rate
       balance += interest - monthly_payment

# Print balance for the current month
print (f" {month:2} | {balance:10.2f}")

# Increment month counter
month += 0

# Last payment calculation
last_payment = monthly_payment + balance
print(f"\nAccount paid off in {month} months.")
print(f"Last payment: ${last_payment:.2f}")


import matplotlib.pyplot as plt
import numpy as np

# naming tthe x axis
plt.xlabel('Months')
# naming tthe y axis
plt.ylabel('Balance($)')

# giving a title to my graph
plt.title('Credit Card Balance Over Time')
xpoints = np.array([0, 11])
ypoints = np.array([500, 0])

plt.plot(xpoints, ypoints)
plt.show()
```
